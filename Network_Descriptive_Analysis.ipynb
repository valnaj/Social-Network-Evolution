### Pre-Release Period

```python
import pickle

# Load the graph object using Python's pickle module
with open("C:\\Users\\Admin\\Downloads\\pre_release_network.gpickle", "rb") as f:
    G1 = pickle.load(f)

# Importing the community module
from networkx.algorithms.community import girvan_newman

# Run the Girvan-Newman community detection algorithm
comp = girvan_newman(G1)

# Find the connected components
node_groups = []
for communities in next(comp):
    node_groups.append(list(communities))

    node_groups.append(communities)
```

```python
print("Number of nodes: ", len(G1.nodes()))
print("Number of communities: ", len(node_groups))
print("Number of communities with size bigger than 2: ", len([i for i in node_groups if len(i) > 2]))
print("Size of communities with size bigger than 2: ")
for i in node_groups:
    if len(i) > 2:
        print("Community size: ", len(i))
```

```python

# Draw the graph with different colors for each cluster
pos = nx.spring_layout(G1)  # Layout for visualization
colors = ['r', 'b', 'g', 'c', 'm', 'y', 'k', 'orange', 'purple', 'lime', 'brown', 'pink', 'teal', 'olive', 'navy', 'maroon', 'gray', 'cyan']  
plt.figure(figsize=(10, 10))  # Adjust figure size as needed
for i, nodes in enumerate(node_groups):
    if len(nodes) > 2: #show only groups of nodes with more than 2 nodes
        nx.draw_networkx_nodes(G1, pos, nodelist=nodes, node_color=colors[i % len(colors)],node_size=200)
        for edge in G1.edges():
            if edge[0] in nodes and edge[1] in nodes:
                nx.draw_networkx_edges(G1, pos, edgelist=[edge]) 
plt.axis('off')
plt.show()
```

## Pre-Release-Network

```python
import networkx as nx
import pickle
import pandas as pd

# Load the directed graph
with open("C:\\Users\\Admin\\Downloads\\pre_release_network.gpickle", "rb") as f:
    G1 = pickle.load(f)

# Convert directed graph G1 to an undirected graph
G1_undirected = G1.to_undirected()

# Compute the core number for each node
core_numbers = nx.core_number(G1_undirected)

# Find all cliques in the undirected graph
all_cliques = list(nx.find_cliques(G1_undirected))
max_clique_size = max(len(clique) for clique in all_cliques)

# Determine the range of k values for cores and cliques
core_k_values = range(1, max(core_numbers.values()) + 1)  # Core numbers start from 1
clique_k_values = range(3, max_clique_size + 1)  # assuming we are interested in cliques of size 3 and above

# Initialize a dictionary to store the binary presence of a node in k-cores and cliques of different sizes
node_combined_presence = {
    node: {**{f"{k}-Core": 0 for k in core_k_values}, **{f"{k}-Clique": 0 for k in clique_k_values}}
    for node in G1_undirected.nodes()
}

# Populate the dictionary with binary values indicating core and clique membership
for node in G1_undirected.nodes():
    node_core_num = core_numbers[node]
    for k in core_k_values:
        if node_core_num >= k:
            node_combined_presence[node][f"{k}-Core"] = 1

for clique in all_cliques:
    k = len(clique)
    if k in clique_k_values:
        for node in clique:
            node_combined_presence[node][f"{k}-Clique"] = 1

# Now, create a pandas DataFrame from the node_combined_presence dictionary
data_for_df = []
for node, presence in node_combined_presence.items():
    row = [node] + [presence[f"{k}-Core"] for k in core_k_values] + [presence[f"{k}-Clique"] for k in clique_k_values]
    data_for_df.append(row)

# Create the DataFrame with appropriate column names
columns = ["Node"] + [f"{k}-Core" for k in core_k_values] + [f"{k}-Clique" for k in clique_k_values]
df = pd.DataFrame(data_for_df, columns=columns)

# Display the DataFrame
print(df)

# Now you have the dataframe `df` which contains your combined core and clique data.

```
